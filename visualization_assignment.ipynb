```python
import matplotlib.pyplot as plt
import numpy as np
```

# Grafico 1 (linea)

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x,y)
```

Graficamos la función seno(x) para los valores del 0 al 10.

# Gráfico 2 (barras)

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x,y)
```

Muestra un gráfico de barras de 4 barras de un array generado por numpy. Cada barra se le asigna una frecuencia que proviene de otro array generado por numpy.

# Gráfico 3 (histograma)

```python
data = np.random.gamma(2, 2, 1000)
plt.hist(data, bins=30)
```

Generamos un histograma de una distribución gamma.

# Gráfico 4 (dispersión)

```python
x = np.random.normal(0, 1, 50)
y = x * 2 + np.random.randn(50)
plt.scatter(x, y, c='red', marker='*')
```

Generamos un gráfico de dispersión de datos que estaban en una línea pero se desplazaron una cantidad aleatoria.

# Gráfico 5 (gráfico circular)

```python
helados = 'Chocolate', 'DDL', 'Banana split', 'Crema'
frecuencia = [15, 30, 45, 10]

plt.pie(frecuencia, labels=helados)
```

Graficamos la cantidad de gente que consumió ciertos gustos de helado en un gráfico de torta
